```python
avg_age = iris["Age"].mean(axis=0)
print(avg_age)
```

```python
iris["Age"].fillna(avg_age,inplace=True)
```

```python
iris["Div"].fillna("A",inplace=True)
```

```python
sns.boxplot(data=iris, x='Age');
```

```python
Q1 = iris['Age'].quantile(0.25)
Q3 = iris['Age'].quantile(0.75)
IQR = Q3 - Q1
```

```python
Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR
```

```python
outliers = (iris['Age'] < Lower_limit) | (iris['Age'] > Upper_limit)
```

```python
print('Detected outliers:')
print(iris[outliers])
```

```python
iris.loc[outliers, 'Age'] = iris.loc[~outliers, 'Age'].median()
```

```python
sns.boxplot(data=iris, x='Age');
```
