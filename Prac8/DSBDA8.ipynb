```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('titanic.csv')
dataset.head()
```

```python
dataset.shape
print(dataset.info())
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna()
```

```python
sns.countplot(data=dataset, x='survived', hue='sex')
plt.title('Survival Count by Sex')
plt.show()
```

```python
sns.countplot(data=dataset, x='survived', hue='class')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survived')
plt.show()
```

```python
sns.scatterplot(data=dataset, x='age', y='fare', hue='survived')
plt.title('Relationship between Age, Fare, and Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.histplot(data=dataset, x='fare', bins=20)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting count of passengers by class
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='class')
plt.title('Passenger Count by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting count of passengers by class
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='class' ,hue='survived')
plt.title('Survival Count of class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=dataset, x='embarked', hue='survived')
plt.title('Survival Count by Embarked Port')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')

plt.show()
```

```python
sns.scatterplot(data=dataset, x='age', y='fare', hue='survived')
plt.title('Relationship between Age, Fare, and Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
female_data = dataset[dataset['sex'] == 'female']

# Plotting scatter plot of Age vs. Fare for females
plt.figure(figsize=(10, 8))
sns.scatterplot(data=female_data, x='age', y='fare', hue='survived')
plt.title('Relationship between Age, Fare (Females Only), and Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
```

```python
survived_data = dataset[dataset['survived'] == 1]

# Plotting scatter plot of Class vs. Age for passengers who survived
plt.figure(figsize=(10, 8))
sns.scatterplot(data=survived_data, x='pclass', y='age')
plt.title('Relationship between Class and Age for Survivors')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```
